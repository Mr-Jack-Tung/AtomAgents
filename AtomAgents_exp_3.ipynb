```python
Task = '''
Explore the effect of Nb concentration on the critical fracture toughness for the following crack systems in NbMo alloys:
(a) [100] crack propagation, [010] crack plane, and [001] crack front direction. 
(b) [1-10] crack propagation, [111] crack plane, and [-1-12] crack front direction. 
(c) [111] crack propagation, [11-2] crack plane, and [-110] crack front direction. 
(d) [1-1-2] crack propagation, [110] crack plane, and [1-11] crack front direction. 
Use NbMo.mtp moment tensor potential (Nb is the first element and Mo is the second element) and consider Nb concentration from 0 to 100 in intervals of 20. 
Also, for surface energy calculations, consider 50 Angstrom as the box size along free surfaces and 20 along other directions.
Finally, plot the results as a function of Nb concentration and discuss them. What is the preferential crack plane in each system?'''
```
